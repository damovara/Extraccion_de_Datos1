_**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**_

```python
#Filtro por comparacion FILTRO 6
filtro6=tabla[tabla["population"] > 800000]
filtro6 #Filtrar en la columna population datos mayores a 800000
```

_**FILTROS BASICOS CON CONETORES LOGICOS**_

```python
#Filtro Y (FILTRO 7) filtrar la población mayor a 800000 y la región de Southern Europe
filtro7=tabla[(tabla["population"] > 800000) & (tabla["region"] == "Southern Europe")]
filtro7
```

```python
#Filtro o (FILTRO 8) filtrar la población mayor a 800000 ó la región de Southern Europe
filtro8=tabla[(tabla["population"] > 800000) | (tabla["region"] == "Southern Europe")]
filtro8
```

```python
#Filtro not (FILTRO 9) no filtrar la población mayor a 800000 y diferente a la región de Southern Europe
filtro9=tabla[~(tabla["population"] >= 800000) & ~ (tabla["region"] == "Southern Europe")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
